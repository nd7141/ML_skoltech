```python
%matplotlib inline
from sklearn.datasets import make_blobs, make_circles
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#### Exercise 1

```python
X, y = make_blobs(1000, 100)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.decomposition import PCA, KernelPCA
pca = PCA(n_components=3)
pca.fit(X)
pca.explained_variance_ratio_
```

```python
pca2 = PCA(2, svd_solver='arpack')
X_pca = pca.fit_transform(X)
plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
```

#### Exercise 2

```python
X, y = make_circles(1000, shuffle=True, noise=0.1, factor=0.3)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)
plt.scatter(X_pca, np.zeros((X_pca.shape[0], 1)), c=y)
pca.explained_variance_ratio_
```

```python
for i in range(10):
    kpca = KernelPCA(n_components=1, kernel='rbf', gamma=i)
    X_kpca = kpca.fit_transform(X)
    plt.scatter(X_kpca, np.zeros((X_kpca.shape[0], 1)), c=y, lw=0, alpha = 0.5)
    plt.title("Gamma={}".format(i))
    plt.show()
```
